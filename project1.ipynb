```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib
```

```python
data = pd.read_csv("cost_revenue_clean.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
x_cor = DataFrame(data, columns=["production_budget_usd"])
y_cor = DataFrame(data, columns={"worldwide_gross_usd"})
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_cor,y_cor)
```

```python
slope=regression.coef_
```

```python
intercept=regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.title("Movie revenue vs production budget",fontsize=16)
plt.scatter(x_cor,y_cor,alpha=0.45)
plt.plot(x_cor,regression.predict(x_cor),color="purple",linewidth=3) # we can also do it this way slope*x + intercept
plt.xlim(0, 450000000)
plt.ylim(0,2500000000)
plt.xlabel("Production budget $",fontsize = 14)
plt.ylabel("Movie Revenue $", fontsize = 14)
```

```python
print(f"the line above has slope{slope}, and y_intercept as {intercept}")
```

# The model is as follows
## $$y = 3.11x -7.236193*10^6$$
###  where y is the movie revenue and x the production budget.

###  for a given x value, the corresponding y value can be gotten by simply substituting the x value in the linear relationshio above

####  the slope of 3.11 indicates that for a production budget of 1 dollar, approximately 3 dollars will be gotten in return(realistic), and the intercept indicates that for a 0 dallar production budget approximately 7230000 will be lost(quite unrealistic) 

```python
regression.score(x_cor,y_cor) # to calculate goodness of fit or how good our model is
```

## a goodness of fit of 55% is not bad but not the best
