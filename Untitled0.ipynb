```python
numbers = [-25.23,-13.897,-9, 28,36,46,49,53,57.43,67.33,70.45]
```

The code below generates outputs base on given inputs and given algorithm

```python
 output = []
def generate_val():
  for number in numbers:
    val = 6.25*number + 30 
    output.append(val)
  return "numbers: {}, output: {}".format(numbers,output)
```

```python
generate_val()
```

# The task here is to use ML techniques to find m and c in the equation : y = mx + c.
# Given the inputset and the outputset generated by the code above.

```python

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_arr = np.array(numbers,dtype=float)
output_arr = np.array(output,dtype=float)
```

```python
for i,number in enumerate(input_arr):
    print("%s is mapped to %s"%(number,output[i]))
```

# Building the model

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

# model compilation

```python
model.compile(loss = "mean_squared_error",optimizer= tf.keras.optimizers.Adam(0.1))
```

# Training the model

```python
history = model.fit(input_arr, output_arr, epochs=1500)
print("Finish training the model")
```



# displaying result



```python
plt.title("Graph of number of loss magnitude against number of epochs values")
plt.xlabel("Number of epochs")
plt.ylabel("Loss magnitude")
plt.plot(history.history["loss"])
```

# Making predictions

```python
pred = model.predict([100.234])
print(pred)
```
