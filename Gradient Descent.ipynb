```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Optimizing a simple cost function:
# $$f(x) = x^2 + x + 1 $$

```python
x_1 = np.linspace(start=-3,stop=3, num=100) #generating data
```

```python
def f(x):
    return x**2 + x + 1
```

```python

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize=16)
plt.ylabel("f(x)",fontsize=16)
plt.plot(x_1,f(x_1),color="purple",linewidth=2)

```

# slope and derivatives

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
plt.title("simple cost function",fontsize = 25)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x_1", fontsize=14)
plt.ylabel("f(x)",fontsize=14)
plt.plot(x_1, f(x_1),color="purple",linewidth=2)


plt.subplot(1,2,2)
plt.grid()
plt.title("Derivative of a cost function", fontsize= 25)
plt.xlim(-0.8,3)
plt.ylim(-2,6)
plt.xlabel("x_1", fontsize=14)
plt.ylabel("df(x)",fontsize=14)
plt.plot(x_1, df(x_1), color="violet",linewidth=3)
```

# gradient descent

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precession = 0.0001
x_list = [new_x]
slope_list = [df(new_x)]
```

```python
for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - gradient * step_multiplier
    step_size = abs(previous_x - gradient*step_multiplier)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size < precession:
        break
```

```python
print("f(x) takes maximum at:", f(new_x))
print("maximum occurs at:", previous_x)
print("the gradient at x is:", df(new_x))
print(f"the loop ran {n} times") 
```

# graphing  the slope and new_x values

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
plt.title("simple cost function",fontsize = 25)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x_1", fontsize=14)
plt.ylabel("f(x)",fontsize=14)
plt.plot(x_1, f(x_1),color="purple",linewidth=2)
values = np.array(x_list)
plt.scatter(x_list,f(values),color="red",s=150,alpha=0.8)


plt.subplot(1,2,2)
plt.grid()
plt.title("Derivative of a cost function", fontsize= 25)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.xlabel("x_1", fontsize=14)
plt.ylabel("df(x)",fontsize=14)
plt.plot(x_1, df(x_1), color="violet",linewidth=3)
plt.scatter(x_list,df(values),color="red",s=150,alpha=0.8)
```
